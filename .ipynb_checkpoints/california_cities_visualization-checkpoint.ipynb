# CALIFORNIA CITIES VISUALIZATION

Import essential libiries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Change matplotlib theme

```python
plt.style.use('seaborn-whitegrid')
```

Read the dataset

```python
data = pd.read_csv("california_cities.csv")
data.head()
```

Extract latd and longd from dataset

```python
latd , longd = data['latd'], data['longd']
```

```python
population,area = data['population_total'],data['area_total_km2']
```

Examine area in california cities

```python
area.describe()
```

```python
area_range = [50,100,200,300,500]
```

Plot latitude and longitude on scatter plot

```python
plt.figure(figsize=(10,6))
plt.scatter(longd ,latd, c = np.log10(population), 
            cmap = 'viridis',s = area, alpha = 0.9)
plt.colorbar(label="log$_{10}${population}")
plt.xlabel('Longitude ')
plt.ylabel('Latitude')
plt.title("California cities: Population and Size Distribution")

for a in area_range:
    plt.scatter([],[],s=a,c='k',label=str(a) + " km$^2$")

plt.legend(labelspacing = 1,title = 'Area')
plt.show()
```
